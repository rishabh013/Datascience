```python
df = pd.read_csv('nyc_taxi_trip_duration.csv')
df
```

```python
df['pickup_datetime'] = pd.to_datetime(df.pickup_datetime)
df['year'] =  df['pickup_datetime'].dt.year
df['month'] = df['pickup_datetime'].dt.month
df['week'] =  df['pickup_datetime'].dt.week
df['day'] =   df['pickup_datetime'].dt.day
df['weekday'] = df['pickup_datetime'].dt.weekday #(day of week)
```

```python
df['hour'] = df['pickup_datetime'].dt.hour
df['minute'] = df['pickup_datetime'].dt.minute
df['minute_of_day'] = df['hour'] * 60 + df['minute']
df['real_hour'] = df['minute_of_day'] / 60
df['is_weekend'] = df['weekday'] > 4
df['is_weekend'] = np.where((df.is_weekend == True),1,0)
```

```python
df['inning'] = df['pickup_datetime'].dt.hour
df['inning'] = np.where((df.inning <= 6),1,df['inning'])
df['inning'] = np.where(((df.inning > 6) & (df.inning <= 12)),2,df['inning'])
df['inning'] = np.where(((df.inning > 12) & (df.inning <= 18)),3,df['inning'])
df['inning'] = np.where(((df.inning > 18) & (df.inning <= 24)),4,df['inning'])
```

```python
df['store_and_fwd_flag'] = df.store_and_fwd_flag.map({'N':0,'Y':1})
```

```python
# Haversine formula for distance
from math import radians, cos, sin, asin, sqrt

def haversine(row):
    lon1 = row['pickup_longitude']
    lat1 = row['pickup_latitude']
    lon2 = row['dropoff_longitude']
    lat2 = row['dropoff_latitude']
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    km = 6369 * c
    return km
```

```python
# Direction formula
def ft_degree(row):
    lon1 = row['pickup_longitude']
    lat1 = row['pickup_latitude']
    lon2 = row['dropoff_longitude']
    lat2 = row['dropoff_latitude']
    km = 6369
    lon_delta_rad = np.radians(lon2 - lon1)
    lat1, lon1, lat2, lon2 = map(np.radians, (lat1, lon1, lat2, lon2))
    y = np.sin(lon_delta_rad) * np.cos(lat2)
    x = np.cos(lat1) * np.sin(lat2) - np.sin(lat1) * np.cos(lat2) * np.cos(lon_delta_rad)
    return np.degrees(np.arctan2(y, x))
```

```python
df['distance'] = df.apply(haversine, axis=1)
df['direction'] = df.apply(ft_degree, axis=1)
```

```python
df.drop(columns = ['id','pickup_datetime','dropoff_datetime','pickup_longitude', 'pickup_latitude',
       'dropoff_longitude', 'dropoff_latitude','hour','minute','year'], inplace = True)
```

### (i) Distribution of categorical variable

```python
def UVA_category(data, var_group):
  '''
  Univariate_Analysis_categorical
  takes a group of variables (category) and plot/print all the value_counts and barplot.
  '''
  # setting figure_size
  size = len(var_group)
  plt.figure(figsize = (5*size,7), dpi = 200)

  # for every variable
  for j,i in enumerate(var_group):
        norm_count = data[i].value_counts(normalize = True)
        n_uni = data[i].nunique()

      #Plotting the variable with every information
        plt.subplot(1,size,j+1)
        sns.barplot(norm_count, norm_count.index , order = norm_count.index)
        plt.xlabel('fraction/percent', fontsize = 20)
        plt.ylabel('{}'.format(i), fontsize = 20)
        plt.title('n_uniques = {} \n value counts \n {};'.format(n_uni,norm_count))
```

```python
def UVA_numeric(data, var_group):
    '''
    Univariate_Analysis_numeric
    takes a group of variables (INTEGER and FLOAT) and plot/print all the descriptives and properties along with KDE.

    Runs a loop: calculate all the descriptives of i(th) variable and plot/print it
    '''

    size = len(var_group)
    plt.figure(figsize = (7*size,3), dpi = 100)

    #looping for each variable
    for j,i in enumerate(var_group):
        # calculating descriptives of variable
        mini = data[i].min()
        maxi = data[i].max()
        ran = data[i].max()-data[i].min()
        mean = data[i].mean()
        median = data[i].median()
        st_dev = data[i].std()
        skew = data[i].skew()
        kurt = data[i].kurtosis()

        # calculating points of standard deviation
        points = mean-st_dev, mean+st_dev

        #Plotting the variable with every information
        plt.subplot(1,size,j+1)
        sns.kdeplot(data[i], shade=True)
        sns.lineplot(points, [0,0], color = 'black', label = "std_dev")
        sns.scatterplot([mini,maxi], [0,0], color = 'orange', label = "min/max")
        sns.scatterplot([mean], [0], color = 'red', label = "mean")
        sns.scatterplot([median], [0], color = 'blue', label = "median")
        plt.xlabel('{}'.format(i), fontsize = 20)
        plt.ylabel('density')
        plt.title('std_dev = {}; kurtosis = {};\nskew = {}; range = {}\nmean = {}; median = {}'.format((round(points[0],2),round(points[1],2)),
                                                                                                       round(kurt,2),
                                                                                                       round(skew,2),
                                                                                                       (round(mini,2),round(maxi,2),round(ran,2)),
                                                                                                       round(mean,2),
                                                                                                       round(median,2)))
```

### Outliers

```python
def UVA_outlier(data, var_group, include_outlier = True):
  '''
  Univariate_Analysis_outlier:
  takes a group of variables (INTEGER and FLOAT) and plot/print boplot and descriptives\n
  Runs a loop: calculate all the descriptives of i(th) variable and plot/print it \n\n

  data : dataframe from which to plot from\n
  var_group : {list} type Group of Continuous variables\n
  include_outlier : {bool} whether to include outliers or not, default = True\n
  '''

  size = len(var_group)
  plt.figure(figsize = (7*size,4), dpi = 100)
  
  #looping for each variable
  for j,i in enumerate(var_group):
    
    # calculating descriptives of variable
    quant25 = data[i].quantile(0.25)
    quant75 = data[i].quantile(0.75)
    IQR = quant75 - quant25
    med = data[i].median()
    whis_low = med-(1.5*IQR)
    whis_high = med+(1.5*IQR)

    # Calculating Number of Outliers
    outlier_high = len(data[i][data[i]>whis_high])
    outlier_low = len(data[i][data[i]<whis_low])

    if include_outlier == True:
      print(include_outlier)
      #Plotting the variable with every information
      plt.subplot(1,size,j+1)
      sns.boxplot(data[i], orient="v")
      plt.ylabel('{}'.format(i))
      plt.title('With Outliers\nIQR = {}; Median = {} \n 2nd,3rd  quartile = {};\n Outlier (low/high) = {} \n'.format(
                                                                                                   round(IQR,2),
                                                                                                   round(med,2),
                                                                                                   (round(quant25,2),round(quant75,2)),
                                                                                                   (outlier_low,outlier_high)
                                                                                                   ))
      
    else:
      # replacing outliers with max/min whisker
      data2 = data[var_group][:]
      data2[i][data2[i]>whis_high] = whis_high+1
      data2[i][data2[i]<whis_low] = whis_low-1
      
      # plotting without outliers
      plt.subplot(1,size,j+1)
      sns.boxplot(data2[i], orient="v")
      plt.ylabel('{}'.format(i))
      plt.title('Without Outliers\nIQR = {}; Median = {} \n 2nd,3rd  quartile = {};\n Outlier (low/high) = {} \n'.format(
                                                                                                   round(IQR,2),
                                                                                                   round(med,2),
                                                                                                   (round(quant25,2),round(quant75,2)),
                                                                                                   (outlier_low,outlier_high)
                                                                                                   ))
```

## Outlier tratment

```python
outlier = ['week','distance']
```

```python
# column should be integer
def capping_flooring(df,colu):
    data = sorted(df[colu].to_list())
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    df[colu]=np.where(df[colu]>upr_bound,upr_bound,df[colu])
    df[colu]=np.where(df[colu]<lwr_bound,upr_bound,df[colu])
    return df
```

```python
for i in outlier:
    df=capping_flooring(df,i)
```

### (i) Z test

*null hypothesis - average trip duration of 2 vendors are equal                                  
*alternate hypothesis - average trip duration of 2 vendors are not equal

* why z-test ?
* To determine if the mean of a population is different from a hypothesized value or if the means of two different populations are significantly different from each other
* when sample standard deviation given
* n >= 30
* continuous variable & categorical variable

```python
import scipy.stats

data_1 = df1[df1['vendor_id']== 1].sample(150)
data_2 = df1[df1['vendor_id']== 2].sample(150)

group1 = data_1["trip_duration"]
group2 = data_2["trip_duration"]

test_stat, p_value  = scipy.stats.ttest_rel(group1, group2)  #Performing a paired t-test

# interpret the results
if p_value > 0.05:
    print("Accept the HO ")
else:
    print("Reject the HO ")
```

### (ii) t-test

*null hypothesis - average trip duration of 2 vendors are equal                                  
*alternate hypothesis - average trip duration of 2 vendors are not equal

* why t-test ?
* To determine if the mean of a population is different from a hypothesized value or if the means of two different populations are significantly different from each other
* when sample standard deviation given
* n < 30
* continuous variable & categorical variable

```python
from scipy import stats

# Select a subset of the data with(continuous variable & categorical variable)

data_1 = df1[df1['vendor_id']== 1].sample(25)
data_2 = df1[df1['vendor_id']== 2].sample(25)

group1 = data_1["trip_duration"]
group2 = data_2["trip_duration"]

# perform the t-test
t_score, p_value = stats.ttest_ind(group1, group2)
print(t_score,"and", p_value)

# interpret the results
if p_value > 0.05:
    print("Accept the HO ")
else:
    print("Reject the HO ")
```

**Inferences**

1. There is a significant difference between the means of two samples.
2. vendor_id_1 and vendor_id_2 both are independent.

### (iii) CHI Sqare test

It is commonly used to determine if there is a significant difference between the expected frequencies and the observed frequencies in one or more categories.(cat-cat) (cat independent and cat dependent)

* p > 0.05  There is any no relationship between the two variables
* p< 0.05   There is sufficient evidence the relation b/w two variables

```python
from scipy.stats import chi2_contingency

sample1=df1[df1['store_and_fwd_flag']==1].sample(50)
sample2=df1[df1['store_and_fwd_flag']==0].sample(50)

sample=pd.concat([sample1,sample2])

# calculate the observed frequencies
observed = pd.crosstab(sample["vendor_id"], sample["store_and_fwd_flag"])

# perform the chi-squared test
chi2, p_value, dof, expected = chi2_contingency(observed)

# interpret the results
if p_value > 0.05:
    print("Accept the HO ")
else:
    print("Reject the HO ")
print("p_value: ", p_value)
```

**Interference**

1. There is a relationship between the vendor_id of a store_and_fwd_flag
