## Bagging

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

* we are seeing above the decison tree that the tree create base on the Tree lavel for the 2 features
* means the entire decison tree created with 2 columns (0,1 columns).

## Random Forest

```python
rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```

* we are seeing above the decison tree that the tree create base on the node lavel for the 2 features
* means any two column migh be come at the node point.
